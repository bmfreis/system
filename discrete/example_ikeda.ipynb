# Ikeda map

Ikeda map is a discrete time dynamical system defined by two difference equations:

$$\begin{array}{ll}
        x[n+1] = 1 + u \; (x[n] \, \cos \left(t[n]\right) -y[n] \, \sin \left(t[n]\right)) \\
        y[n+1] =     u \; (x[n] \, \sin \left(t[n]\right) +y[n] \, \cos \left(t[n]\right))
    \end{array} \; , $$
    
where $u \in \mathbb{R}_{+}^{*}$ is a parameter and $t[n] = 0.4-\dfrac{6}{1+x[n]^{2}+y[n]^{2}}$.




```python
import discrete
import numpy as np
import matplotlib.pyplot as plt
```

## Periodic example

```python
t, x, y = discrete.ikeda_map(t = 5100, u = 0.62, xn = 0.1, yn = 0.7)
ti = 5000
tf = 5100
```

```python
# Plotting time series
plt.figure(num = None, figsize = ((12,6)), dpi = 100)
ax1 = plt.subplot(2,1,1)
plt.plot(t[ti:tf], x[ti:tf], '.k')
plt.ylabel(r"$x[n]$")
plt.setp(ax1.get_xticklabels(), visible=False)
ax2 = plt.subplot(2,1,2,sharex=ax1)
plt.plot(t[ti:tf], y[ti:tf], '.k')
plt.ylabel(r"$y[n]$")
plt.xlabel(r"$n$")
plt.xlim(t[ti-1], t[tf])
plt.show()
```

```python
# Plotting attractor
fig = plt.figure(num = None, figsize = ((6,6)), dpi = 100)
ax  = fig.add_subplot(1, 1, 1)
ax.plot(x[ti:tf], y[ti:tf], '.k')
ax.grid( True )
ax.set_xlabel(r"$x[n]$", labelpad = 10)
ax.set_ylabel(r"$y[n]$", labelpad = 10)
plt.show()
```

## Chaotic example

```python
t, x, y = discrete.ikeda_map(t = 10000, u = 0.9, xn = 0.5, yn = 0.4)
ti = 5000
tf = 10000
```

```python
# Plotting time series
plt.figure(num = None, figsize = ((12,6)), dpi = 100)
ax1 = plt.subplot(2,1,1)
plt.plot(t[ti:tf], x[ti:tf], ',k')
plt.ylabel(r"$x[n]$")
plt.setp(ax1.get_xticklabels(), visible=False)
ax2 = plt.subplot(2,1,2,sharex=ax1)
plt.plot(t[ti:tf], y[ti:tf], ',k')
plt.ylabel(r"$y[n]$")
plt.xlabel(r"$n$")
plt.xlim(t[ti-1], t[tf-1])
plt.show()
```

```python
# Plotting attractor
fig = plt.figure(num = None, figsize = ((6,6)), dpi = 100)
ax  = fig.add_subplot(1, 1, 1)
ax.plot(x[ti:tf], y[ti:tf], ',k')
ax.grid( True )
ax.set_xlabel(r"$x[n]$", labelpad = 10)
ax.set_ylabel(r"$y[n]$", labelpad = 10)
plt.show()
```

## Bifurcation diagram

```python
u_min, u_max, du = 0.0, 0.9, 0.001
u_values = np.arange(u_min, u_max+du, du)
ti = 1800
tf = 2000

fig = plt.figure(num = None, figsize=(10,10), dpi = 100)
ax  = fig.add_subplot(2, 1, 1)
bx  = fig.add_subplot(2, 1, 2, sharex = ax)
for u in u_values:
    t, x, y = discrete.ikeda_map(t = 2000, u = u, xn = 0.5, yn = 0.4)
    ax.plot([u]*200, x[ti:tf], ',k')
    bx.plot([u]*200, y[ti:tf], ',k')
    del t, x, y
ax.set_ylabel(r"$x$", labelpad = 10)
bx.set_ylabel(r"$y$", labelpad = 10)
bx.set_xlabel(r"$u$", labelpad = 10)
bx.set_xlim(u_min, u_max)
plt.setp(ax.get_xticklabels(), visible=False)
plt.show()
```
