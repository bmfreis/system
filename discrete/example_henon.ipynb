# Hénon map

The Hénon map [1] is a discrete time dynamical system defined by two difference equations:

$$\begin{array}{ll}
        x[n+1] \; = \; 1 - \alpha \: x[n]^{2} + y[n] \\
        y[n+1] \; = \; \beta \: x[n]
    \end{array} \; , $$
    
where $\alpha > \frac{\left(\, b-1 \, \right)^{2}}{4}$ and $b \in (0,1)$ are parameters.

```python
import discrete
import numpy as np
import matplotlib.pyplot as plt
```

## Periodic example

```python
t, x, y = discrete.henon_map(n = 5100, a = 1.25, b = 0.3, x0 = 0.5, y0 = 0.4)
ti = 5000
tf = 5100
```

```python
# Plotting time series
plt.figure(num = None, figsize = ((12,6)), dpi = 100)
ax1 = plt.subplot(2,1,1)
plt.plot(t[ti:tf], x[ti:tf], '.k')
plt.ylabel(r"$x[n]$")
plt.setp(ax1.get_xticklabels(), visible=False)
ax2 = plt.subplot(2,1,2,sharex=ax1)
plt.plot(t[ti:tf], y[ti:tf], '.k')
plt.ylabel(r"$y[n]$")
plt.xlabel(r"$n$")
plt.xlim(t[ti-1], t[tf])
plt.show()
```

```python
# Plotting attractor
fig = plt.figure(num = None, figsize = ((6,6)), dpi = 100)
ax  = fig.add_subplot(1, 1, 1)
ax.plot(x[ti:tf], y[ti:tf], '.k')
ax.grid( True )
ax.set_xlabel(r"$x[n]$", labelpad = 10)
ax.set_ylabel(r"$y[n]$", labelpad = 10)
plt.show()
```

## Chaotic example

```python
t, x, y = discrete.henon_map(n = 10000, a = 1.4, b = 0.3, x0 = 0.5, y0 = 0.4)
ti = 5000
tf = 10000
```

```python
# Plotting time series
plt.figure(num = None, figsize = ((12,6)), dpi = 100)
ax1 = plt.subplot(2,1,1)
plt.plot(t[ti:tf], x[ti:tf], ',k')
plt.ylabel(r"$x[n]$")
plt.setp(ax1.get_xticklabels(), visible=False)
ax2 = plt.subplot(2,1,2,sharex=ax1)
plt.plot(t[ti:tf], y[ti:tf], ',k')
plt.ylabel(r"$y[n]$")
plt.xlabel(r"$n$")
plt.xlim(t[ti-1], t[tf-1])
plt.show()
```

```python
# Plotting attractor
fig = plt.figure(num = None, figsize = ((6,6)), dpi = 100)
ax  = fig.add_subplot(1, 1, 1)
ax.plot(x[ti:tf], y[ti:tf], ',k')
ax.grid( True )
ax.set_xlabel(r"$x[n]$", labelpad = 10)
ax.set_ylabel(r"$y[n]$", labelpad = 10)
plt.show()
```

## Bifurcation diagram

```python
a_min, a_max, da = 0.0005, 1.42, 0.0005
a_values = np.arange(a_min, a_max+da, da)
ti = 1800
tf = 2000
dt = tf-ti

fig = plt.figure(num = None, figsize=(10,10), dpi = 100)
ax  = fig.add_subplot(2, 1, 1)
bx  = fig.add_subplot(2, 1, 2, sharex = ax)
for a in a_values:
    t, x, y = discrete.henon_map(n = 2000, a = a, b = 0.3, x0 = 0.5, y0 = 0.4)
    ax.plot([a]*200, x[ti:tf], ',k')
    bx.plot([a]*200, y[ti:tf], ',k')
    del t, x, y
ax.set_ylabel(r"$x$", labelpad = 10)
bx.set_ylabel(r"$y$", labelpad = 10)
bx.set_xlabel(r"$a$", labelpad = 10)
bx.set_xlim(a_min, a_max)
plt.setp(ax.get_xticklabels(), visible=False)
plt.show()
```
