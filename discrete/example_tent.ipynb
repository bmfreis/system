# Tent map

Tent map is a discrete time dynamical system defined by:

$$x[n+1] = \mu \: \min\left( \, x[n],\,1-x[n] \, \right),$$
    	
where $x \in [0,1]$ and $\mu \in [0,2]$. 

```python
import discrete
import numpy as np
import matplotlib.pyplot as plt
```

## Periodic example

```python
t, x = discrete.tent_map(n = 1000, mu = 1.1, x0 = 0.2)
ti = 900
tf = 1000
```

```python
# Plotting time series
plt.figure(num = None, figsize = ((9,3)), dpi = 100)
plt.plot(t[ti:tf], x[ti:tf], '.k')
plt.ylabel(r"$x[n]$")
plt.xlabel(r"$n$")
plt.xlim(t[ti-1], t[tf])
plt.show()
```

```python
# Plotting first return
plt.figure(num = None, figsize = ((6,6)), dpi = 100)
plt.plot(x[ti:tf-1:1], x[ti+1:tf:1], '.k')
plt.grid( True )
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel(r"$x[n]$")
plt.ylabel(r"$x[n+1]$")
plt.xticks([0.00, 0.25, 0.50, 0.75, 1.00])
plt.yticks([0.00, 0.25, 0.50, 0.75, 1.00])
plt.show()
```

## Chaotic example

```python
t, x = discrete.tent_map(n = 1000, mu = 1.6, x0 = 0.2)
ti = 500
tf = 1000
```

```python
# Plotting time series
plt.figure(num = None, figsize = ((9,3)), dpi = 100)
plt.plot(t[ti:tf], x[ti:tf], '.k')
plt.ylabel(r"$x[n]$")
plt.xlabel(r"$n$")
plt.xlim(t[ti-1], t[tf])
plt.show()
```

```python
# Plotting first return
plt.figure(num = None, figsize = ((6,6)), dpi = 100)
plt.plot(x[ti:tf-1:1], x[ti+1:tf:1], '.k')
plt.grid( True )
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel(r"$x[n]$")
plt.ylabel(r"$x[n+1]$")
plt.xticks([0.00, 0.25, 0.50, 0.75, 1.00])
plt.yticks([0.00, 0.25, 0.50, 0.75, 1.00])
plt.show()
```

## Bifurcation diagram

```python
mu_min, mu_max, dmu = 0.001, 2.0, 0.001
mu_values = np.arange(mu_min, mu_max+dmu, dmu)
ti = 1850
tf = 2000
dt = tf-ti

plt.figure(num = None, figsize=(10,10), dpi = 100)
for mu in mu_values:
    t, x = discrete.tent_map(n = 2000, mu = mu, x0 = 0.2)
    plt.plot([mu]*dt, x[ti:tf], ',k')
    del t, x
plt.ylabel(r"$x$", labelpad = 10)
plt.xlabel(r"$\mu$", labelpad = 10)
plt.xlim(mu_min, mu_max)
plt.show()
```
