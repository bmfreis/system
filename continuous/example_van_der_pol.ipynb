# Van der Pol oscillator

Van der Pol oscillator is a continuous time dynamical system defined by second order differential equation:

$$\ddot{x} - \mu \left( 1 - x^{2} \right) \dot{x} + x = 0,$$

where $x$ is the position coordinate and $\mu$ is a parameter indicating the damping. Using the transformation $y = \dot{x}$, the system can be defined by two first order differential equation:

$$\begin{array}{ll}
        \dot{x} & = y \\
        \dot{y} & = \mu \left( 1 - x^{2} \right) y - x
    \end{array} \; . $$

```python
import continuous
import matplotlib.pyplot as plt
```

## Example with damping $\mu = 0.2$

```python
t, x, y = continuous.van_der_pol_system(t = 300, dt = 0.01, mu = 0.2, xi = 0.1, yi = 0.2)
ti = 0
tf = 30000
```

```python
# Plotting time series
plt.figure(num = None, figsize = ((6,6)), dpi = 100)
ax1 = plt.subplot(2,1,1)
plt.plot(t[ti:tf], x[ti:tf], '-k')
plt.ylabel(r"$x(t)$")
plt.setp(ax1.get_xticklabels(), visible=False)
ax2 = plt.subplot(2,1,2,sharex=ax1)
plt.plot(t[ti:tf], y[ti:tf], '-k')
plt.ylabel(r"$y(t)$")
plt.xlabel(r"$t$")
plt.xlim(t[ti], t[tf])
plt.show()
```

```python
# Plotting attractor
fig = plt.figure(num = None, figsize = ((6,6)), dpi = 100)
ax  = fig.add_subplot(1, 1, 1)
ax.plot(x[ti:tf], y[ti:tf], linewidth = 1, color = 'k')
ax.grid( True )
ax.set_xlabel(r"$x(t)$", labelpad = 10)
ax.set_ylabel(r"$y(t)$", labelpad = 10)
plt.show()
```
