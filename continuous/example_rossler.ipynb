# Rössler system

Rössler system [1] is a continuous time dynamical system defined by three non-linear ordinary differential equations:

$$\begin{array}{ll}
        \dot{x} & = -y-z \\
        \dot{y} & = x+ay \\
        \dot{z} & = b+z(x-c)
    \end{array} \; , $$
    
    
where $(x,y,z) \in \mathbb{R}^{3}$ are dynamical variables defining the phase space and $(a,b,c) \in \mathbb{R}^{3}$ are parameters.

```python
import continuous
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Periodic example

```python
t, x, y, z = continuous.rossler_system(t = 300, dt = 0.01, a = 0.35, b = 2., c = 4., xi = 1., yi = 2., zi = 3.)
ti = 10000
tf = 30000
```

```python
# Plotting time series
plt.figure(num = None, figsize = ((6,6)), dpi = 100)
ax1 = plt.subplot(3,1,1)
plt.plot(t[ti:tf], x[ti:tf], '-k')
plt.ylabel(r"$x(t)$")
plt.setp(ax1.get_xticklabels(), visible=False)
ax2 = plt.subplot(3,1,2,sharex=ax1)
plt.plot(t[ti:tf], y[ti:tf], '-k')
plt.ylabel(r"$y(t)$")
plt.setp(ax2.get_xticklabels(), visible=False)
ax3 = plt.subplot(3,1,3,sharex=ax1)
plt.plot(t[ti:tf], z[ti:tf], '-k')
plt.ylabel(r"$z(t)$")
plt.xlabel(r"$t$")
plt.xlim(t[ti-1], t[tf-1])
plt.show()
```

```python
# Plotting attractor
fig = plt.figure(num = None, figsize = ((6,6)), dpi = 100)
ax  = fig.add_subplot(1, 1, 1, projection = '3d')
ax.plot(x[ti:tf], y[ti:tf], z[ti:tf], linewidth = 1, color = 'k')
ax.grid( True )
ax.set_xlabel(r"$x(t)$", labelpad = 10)
ax.set_ylabel(r"$y(t)$", labelpad = 10)
ax.set_zlabel(r"$z(t)$", labelpad = 10)
ax.view_init(azim = -70)
plt.show()
```

```python
# Plotting x-y, x-z and y-z projections
fig = plt.figure(num = None, figsize = ((12,4)), dpi = 100)
ax  = fig.add_subplot(1, 3, 1)
ax.plot(x[ti:tf], y[ti:tf], linewidth = 1, color = 'k')
ax.set_xlabel(r"$x(t)$", labelpad = 10)
ax.set_ylabel(r"$y(t)$", labelpad = 10)
bx  = fig.add_subplot(1, 3, 2)
bx.plot(x[ti:tf], z[ti:tf], linewidth = 1, color = 'k')
bx.set_xlabel(r"$x(t)$", labelpad = 10)
bx.set_ylabel(r"$z(t)$", labelpad = 10)
cx  = fig.add_subplot(1, 3, 3)
cx.plot(y[ti:tf], z[ti:tf], linewidth = 1, color = 'k')
cx.set_xlabel(r"$y(t)$", labelpad = 10)
cx.set_ylabel(r"$z(t)$", labelpad = 10)
plt.subplots_adjust(wspace = 0.3)
plt.show()
```

```python
# Mapping local maxima
peaks_x = []
peaks_y = []
peaks_z = []
for u in range (ti,len(z)-1):
    if (x[u]>x[u-1]) and (x[u]>x[u+1]):
        peaks_x.append(u)
    if (y[u]>y[u-1]) and (y[u]>y[u+1]):
        peaks_y.append(u)        
    if (z[u]>z[u-1]) and (z[u]>z[u+1]):
        peaks_z.append(u)

fig = plt.figure(num = None, figsize = ((12,4)), dpi = 100)
ax  = fig.add_subplot(1, 3, 1)
ax.plot(x[peaks_x[0:-1]], x[peaks_x[1:]], '*', color = 'k')
ax.set_xlabel(r"$x(t)$", labelpad = 10)
ax.set_ylabel(r"$x(t+1)$", labelpad = 10)
bx  = fig.add_subplot(1, 3, 2)
bx.plot(y[peaks_y[0:-1]], y[peaks_y[1:]], '*', color = 'k')
bx.set_xlabel(r"$y(t)$", labelpad = 10)
bx.set_ylabel(r"$y(t+1)$", labelpad = 10)
cx  = fig.add_subplot(1, 3, 3)
cx.plot(z[peaks_z[0:-1]], z[peaks_z[1:]], '*', color = 'k')
cx.set_xlabel(r"$z(t)$", labelpad = 10)
cx.set_ylabel(r"$z(t+1)$", labelpad = 10)
plt.subplots_adjust(wspace = 0.3)
plt.show()
```
